```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("C:/Projects/help_crosswalk/Datasets/Demo_images/crop_stripe.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

canny = cv2.Canny(img, 150, 255)
H, W = canny.shape[:2]

erode_size = int(H / 15)
erode_structure = cv2.getStructuringElement(cv2.MORPH_RECT, (erode_size, 1))
erode = cv2.morphologyEx(canny, cv2.MORPH_CLOSE, erode_structure)

plt.figure()
plt.imshow(canny, cmap = "gray")
plt.figure()
plt.imshow(erode, cmap = "gray")
```
